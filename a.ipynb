```python
import matplotlib.pyplot as plt

# Define your lists of numerical values
list1 = [1, 2, 3, 4, 5]
list2 = [2, 3, 4, 5, 6]
list3 = [3, 4, 5, 6, 7]
list4 = [4, 5, 6, 7, 8]
list5 = [5, 6, 7, 8, 9]

# Plot each list with a different color
plt.plot(list1, color='red', label='List 1')
plt.plot(list2, color='blue', label='List 2')
plt.plot(list3, color='green', label='List 3')
plt.plot(list4, color='orange', label='List 4')
plt.plot(list5, color='purple', label='List 5')

# Add labels and legend
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Multiple Lists Plot')
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Step 3: Create data
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 9, 16, 25]

# Step 4: Plot the data
plt.plot(x, y1, marker='o', linestyle='-', color='b')
plt.plot(x, y2, marker='s', linestyle='--', color='r')

# Step 5: Add labels to the lines
for i, txt in enumerate(y1):
    plt.text(x[i], y1[i], f'aenvv', ha='right', fontsize=8, color='b')

for i, txt in enumerate(y2):
    plt.text(x[i], y2[i], f'({x[i]}, {y2[i]})', ha='right', fontsize=8, color='r')

# Step 6: Customize the plot
plt.title('My Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)

# Step 7: Save the plot to a file
plt.savefig('my_plot_with_labels_on_lines.png')

# Step 8: Show the plot (optional)
plt.show()
```
